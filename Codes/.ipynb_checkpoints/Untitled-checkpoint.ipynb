```python
lc2['created_at'] = pd.to_datetime(lc2['created_at'])
lc3['created_at'] = pd.to_datetime(lc3['created_at'])
```

```python
LC2 = pd.DataFrame(columns = ['created_at', 'PM10_2', 'PM2.5_2', 'RH_2', 'Temperature_2', 'CO_2', 'NO2_2', 'NH3_2'])
LC3 = pd.DataFrame(columns = ['created_at', 'PM10_3', 'PM2.5_3', 'RH_3', 'Temperature_3', 'CO_3', 'NO2_3', 'NH3_3' ])
```

```python
pm10 = []
pm25 = []
rh = []
temp = []
co = []
no2 = []
nh3 = []
for i in range(len(lc2)):
    if(len(pm10) == 0):
        time = lc2['created_at'][i]
        pm10.append(lc2['PM10_2'][i])
        pm25.append(lc2['PM2.5_2'][i])
        rh.append(lc2['RH_2'][i])
        temp.append(lc2['Temp_2'][i])
        co.append(lc2['CO_2'])
        no2.append(lc2['NO2_2'])
        nh3.append(lc2['NH3_2'])

    else:

        prev = time - pd.to_timedelta(time.time().minute % 15, unit='m') - pd.to_timedelta(time.time().second, unit='s')
        current_time = lc2['created_at'][i] 
        convert_time = current_time - pd.to_timedelta(current_time.time().minute % 15, unit = 'm') - pd.to_timedelta(current_time.time().second, unit='s')

        if(prev == convert_time):
            time = lc2['created_at'][i]
            pm10.append(lc2['PM10_2'][i])
            pm25.append(lc2['PM2.5_2'][i])
            rh.append(lc2['RH_2'][i])
            temp.append(lc2['Temp_2'][i])
            co.append(lc2['CO_2'])
            no2.append(lc2['NO2_2'])
            nh3.append(lc2['NH3_2'])
    

        else:

            avg_pm10 = round(mean(pm10), 2)
            avg_pm25 = round(mean(pm25), 2)
            avg_rh = round(mean(rh), 2)
            avg_temp = round(mean(temp), 2)
            avg_co = mean(co)
            avg_no2 = mean(no2)
            avg_nh3 = mean(nh3)

            time = time - pd.to_timedelta(time.time().minute % 1,unit='m') - pd.to_timedelta(time.time().second, unit='s')
            data = {'created_at': time, 'PM10_2' : avg_pm10, 'PM2.5_2': avg_pm25, 'RH_2': avg_rh, 'Temperature_2': avg_temp, 'CO_2': avg_co, 'NO2_2': avg_no2, 'NH3_2': avg_nh3}

            LC2 = LC2.append(data, ignore_index=True)

            pm10.clear()
            pm25.clear()
            rh.clear()
            temp.clear()
            co.clear()
            no2.clear()
            nh3.clear()
            
    
            pm10.append(lc2['PM10_2'][i])
            pm25.append(lc2['PM2.5_2'][i])
            rh.append(lc2['RH_2'][i])
            temp.append(lc2['Temp_2'][i])
            co.append(lc2['CO_2'][i])
            no2.append(lc2['NO2_2'][i])
            nh3.append(lc2['NH3_2'][i])
            
    

            time = lc2['created_at'][i]
```
